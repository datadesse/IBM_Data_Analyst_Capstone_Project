## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

df['ConvertedComp'].plot(kind='box', vert=False, figsize=(10, 4))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here

# first quartiles 25% Q1 and  third quartiles 75% Q3.
# IQR (Inter Quartile Range)

Q1 = 26868 
Q3 = 100000
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here

lower_bounds = Q1 -(1.5*IQR)
upper_bounds= Q3 + (1.5*IQR)

print('Lower_bounds = ', lower_bounds)
print( 'Upper_bounds = ', upper_bounds)
```

#### Lower_bound is  -82830 and upper bound is 209698, so anything outside of these values is an outlier.

```python
# No outliers less than  -82830 
len(df[df['ConvertedComp']<  -82830])
```

```python
len(df[df['ConvertedComp'] > 209698])
```

```python
outliers = (df['ConvertedComp']<  -82830) | (df['ConvertedComp'] > 209698)
df[outliers].head(2)
```

```python
df[outliers]['ConvertedComp'].sort_values( ascending=True)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
len(outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_without_outliers = df[(df['ConvertedComp']>= -82830) & (df['ConvertedComp'] <= 209698)]
```

```python
df_without_outliers['ConvertedComp'].max(), df_without_outliers['ConvertedComp'].min()
```

```python
df_without_outliers['ConvertedComp'].median()
```

```python
df_without_outliers['ConvertedComp'].mean()
```

## Compare the data after removing outliers 

```python
#plt.ticklabel_format(style='plain')
plt.figure(figsize=(20,6))

plt.subplot(2,2,1)
plt.ticklabel_format(style='plain')
sns.boxplot(df['ConvertedComp'], orient= 'h').set(title='With Outliers')

plt.subplot(2,2,3)
sns.boxplot(df_without_outliers['ConvertedComp'], orient= 'h').set(title='WithOut Outliers')

plt.show()
```
